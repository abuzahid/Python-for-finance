```python
#importing pakages
import datetime as dt
import matplotlib.pyplot as plt
import pandas as pd
from matplotlib import style
import pandas_datareader.data as web
```

```python
style.use('ggplot')
```

```python
start= dt.datetime(2015,1,1)
end= dt.datetime(2018,12,31)
```

```python
#Collecting data from web
df= web.DataReader('FB', 'yahoo', start, end)
df.tail()
```

```python
#saving data to csv
df.to_csv('FB.csv')
```

```python
#Reading data from csv
fb= pd.read_csv('FB.csv')
```

```python
#plotting 1 coloumn to see the behavour 
fb['Adj Close'].plot()
plt.show()
```

```python
#Creating 100 period EMA(exponential moving average)
fb['100ma']= fb['Adj Close'].rolling(window=100).mean()
```

```python
#plotting the moving average
fb['100ma'].plot()
plt.show()
```

```python
#Plotting price and moving average together 
ax1= plt.subplot2grid((6,1),(0,0), rowspan=5, colspan=1)
ax2= plt.subplot2grid((6,1),(5,0), rowspan=1, colspan=1, sharex= ax1)
ax1.plot(fb.index, fb['Adj Close'])
ax1.plot(fb.index, fb['100ma'])
ax2.bar(fb.index, fb['Volume'])
plt.show()
```
